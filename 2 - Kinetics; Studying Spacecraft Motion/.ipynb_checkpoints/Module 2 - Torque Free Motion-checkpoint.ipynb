# Week 2 - Torque-Free Motion

Week 2 explores the dynamics of single and dual rigid body systems under torque-free conditions, a critical scenario in spacecraft dynamics when no external torques act on the system. The module examines large-scale tumbling motions through the use of **polhode plots**, which visualize the angular velocity motion in the body frame. Analytical rate solutions are derived for both axi-symmetric and general spacecraft shapes, providing insights into how shape and mass distribution affect motion.

The module concludes with an analysis of the **dual-spinner dynamical system**, a practical example of how gyroscopic effects can be leveraged to stabilize a principal axis spin. This is particularly relevant for spacecraft with spinning components like reaction wheels or flywheels, where stability is key to maintaining attitude control.

Understanding torque-free motion is essential for characterizing the natural dynamics of spacecraft, especially during phases where control torques are intentionally minimized or unavailable (e.g., during thruster-free coasting or detumbling).

<ins>**Learning Objectives**</ins>

- Predict and determine torque-free motion equilibria and their associated stabilities
- Describe the torque-free motion of different rigid bodies in space
- Develop equations of motion for rigid bodies with multiple spinning components
- Apply static stability conditions for dual-spinner configurations

```python
import numpy as np
import plotly.graph_objects as go

# =============================================================================
# USER-CONTROLLED PARAMETERS
# =============================================================================

# Moments of inertia (principal axes)
I1 = 3.0
I2 = 2.0
I3 = 1.0

# Angular momentum magnitude (radius of momentum sphere)
H_mag = 5

# Rotational kinetic energy
E_rot = 7

# Sizing for the wireframe lines
N_sphere_wire = 12   # "longitude/latitude" lines for sphere
N_ellip_wire  = 12   # "longitude/latitude" lines for ellipsoid

# Resolution for surfaces
N_sphere_surf   = 60
N_ellip_surf    = 60

# Opacities for surfaces
sphere_opacity = 0.4
ellip_opacity  = 0.4

# Intersection toggle
SHOW_INTERSECTION   = True
INTERSECTION_COLOR  = 'red'
INTERSECTION_SIZE   = 4
INTERSECTION_TOL    = 0.02

# =============================================================================
# 1. GENERATE DATA FOR SURFACES
# =============================================================================

# -----------------------------------------------------------------------------
# 1A. Momentum Sphere surface
#     Equation in H-space: H1^2 + H2^2 + H3^2 = H_mag^2
# -----------------------------------------------------------------------------
phi_s   = np.linspace(0,   np.pi,   N_sphere_surf)
theta_s = np.linspace(0, 2*np.pi,   N_sphere_surf)
phi_s, theta_s = np.meshgrid(phi_s, theta_s)

H1_sphere_surf = H_mag * np.sin(phi_s) * np.cos(theta_s)
H2_sphere_surf = H_mag * np.sin(phi_s) * np.sin(theta_s)
H3_sphere_surf = H_mag * np.cos(phi_s)

# -----------------------------------------------------------------------------
# 1B. Energy Ellipsoid surface
#     (H1^2)/(2I1) + (H2^2)/(2I2) + (H3^2)/(2I3) = E_rot
#     => H1^2/I1 + H2^2/I2 + H3^2/I3 = 2 E_rot
#     We'll param similarly to a sphere but scale in each axis direction.
# -----------------------------------------------------------------------------
u_e   = np.linspace(0,   np.pi,   N_ellip_surf)
v_e   = np.linspace(0, 2*np.pi,   N_ellip_surf)
u_e, v_e = np.meshgrid(u_e, v_e)

A1 = np.sqrt(2.0 * E_rot * I1)  # semi-axis along H1
A2 = np.sqrt(2.0 * E_rot * I2)  # semi-axis along H2
A3 = np.sqrt(2.0 * E_rot * I3)  # semi-axis along H3

H1_ellip_surf = A1 * np.sin(u_e) * np.cos(v_e)
H2_ellip_surf = A2 * np.sin(u_e) * np.sin(v_e)
H3_ellip_surf = A3 * np.cos(u_e)

# =============================================================================
# 2. GENERATE WIREFRAME LINES
#    We'll overlay black “latitude/longitude” lines for both sphere & ellipsoid.
# =============================================================================

def wireframe_sphere(H_mag, n_meridians=12, n_parallels=12, color='black', width=1):
    """
    Returns a list of Scatter3d objects for wireframe lines on a sphere of radius H_mag.
    """
    traces = []
    
    # 2A. "Longitude" lines: fix theta, vary phi
    thetas = np.linspace(0, 2*np.pi, n_meridians, endpoint=False)
    phi_line = np.linspace(0, np.pi, 40)
    for th in thetas:
        x = H_mag * np.sin(phi_line) * np.cos(th)
        y = H_mag * np.sin(phi_line) * np.sin(th)
        z = H_mag * np.cos(phi_line)
        traces.append(go.Scatter3d(
            x=x, y=y, z=z,
            mode='lines',
            line=dict(color=color, width=width),
            showlegend=False
        ))
    
    # 2B. "Latitude" lines: fix phi, vary theta
    phis = np.linspace(0, np.pi, n_parallels+2)[1:-1]  # skip poles
    theta_line = np.linspace(0, 2*np.pi, 40)
    for phi in phis:
        x = H_mag * np.sin(phi) * np.cos(theta_line)
        y = H_mag * np.sin(phi) * np.sin(theta_line)
        z = H_mag * np.cos(phi) * np.ones_like(theta_line)
        traces.append(go.Scatter3d(
            x=x, y=y, z=z,
            mode='lines',
            line=dict(color=color, width=width),
            showlegend=False
        ))
    return traces

def wireframe_ellipsoid(A1, A2, A3, n_meridians=12, n_parallels=12, color='black', width=1):
    """
    Returns a list of Scatter3d objects for wireframe lines on an ellipsoid:
    (H1^2/A1^2) + (H2^2/A2^2) + (H3^2/A3^2) = 1, 
    where A1, A2, A3 are the semi-axes.
    """
    traces = []
    
    # 2A. "Longitude" lines
    v_vals = np.linspace(0, 2*np.pi, n_meridians, endpoint=False)
    u_line = np.linspace(0, np.pi, 40)
    for v in v_vals:
        H1 = A1 * np.sin(u_line) * np.cos(v)
        H2 = A2 * np.sin(u_line) * np.sin(v)
        H3 = A3 * np.cos(u_line)
        traces.append(go.Scatter3d(
            x=H1, y=H2, z=H3,
            mode='lines',
            line=dict(color=color, width=width),
            showlegend=False
        ))
        
    # 2B. "Latitude" lines
    u_vals = np.linspace(0, np.pi, n_parallels+2)[1:-1]
    v_line = np.linspace(0, 2*np.pi, 40)
    for u in u_vals:
        H1 = A1 * np.sin(u) * np.cos(v_line)
        H2 = A2 * np.sin(u) * np.sin(v_line)
        H3 = A3 * np.cos(u) * np.ones_like(v_line)
        traces.append(go.Scatter3d(
            x=H1, y=H2, z=H3,
            mode='lines',
            line=dict(color=color, width=width),
            showlegend=False
        ))
    return traces

# Create wireframes
sphere_wire_traces = wireframe_sphere(H_mag, 
                                      n_meridians=N_sphere_wire, 
                                      n_parallels=N_sphere_wire, 
                                      color='black', width=1)
ellip_wire_traces  = wireframe_ellipsoid(A1, A2, A3, 
                                         n_meridians=N_ellip_wire, 
                                         n_parallels=N_ellip_wire, 
                                         color='black', width=1)

# =============================================================================
# 3. (OPTIONAL) INTERSECTION POINTS
#    Check which points on the sphere surface also satisfy the ellipsoid eqn.
# =============================================================================
H1_int = []
H2_int = []
H3_int = []
if SHOW_INTERSECTION:
    # We'll re-use (phi_s, theta_s) from the sphere surface
    # Evaluate the ellipsoid eqn for each sphere point:
    #   E_check = (H1^2)/(2I1) + (H2^2)/(2I2) + (H3^2)/(2I3)
    #   compare with E_rot
    E_check = ((H1_sphere_surf**2)/(2.0*I1) + 
               (H2_sphere_surf**2)/(2.0*I2) + 
               (H3_sphere_surf**2)/(2.0*I3))
    mask = np.isclose(E_check, E_rot, atol=INTERSECTION_TOL)
    
    H1_int = H1_sphere_surf[mask]
    H2_int = H2_sphere_surf[mask]
    H3_int = H3_sphere_surf[mask]

# =============================================================================
# 4. BUILD THE FIGURE
# =============================================================================
fig = go.Figure()

# 4A. Add momentum sphere as a partially transparent surface
fig.add_trace(go.Surface(
    x=H1_sphere_surf,
    y=H2_sphere_surf,
    z=H3_sphere_surf,
    opacity=sphere_opacity,
    colorscale='Greens',
    showscale=False,
    name='Momentum Sphere'
))

# 4B. Add energy ellipsoid as a partially transparent surface
fig.add_trace(go.Surface(
    x=H1_ellip_surf,
    y=H2_ellip_surf,
    z=H3_ellip_surf,
    opacity=ellip_opacity,
    colorscale='Blues',
    showscale=False,
    name='Energy Ellipsoid'
))

# 4C. Overlay wireframe lines (black)
for tr in sphere_wire_traces:
    fig.add_trace(tr)
for tr in ellip_wire_traces:
    fig.add_trace(tr)

# 4D. (Optional) Intersection points in red
if SHOW_INTERSECTION and len(H1_int) > 0:
    fig.add_trace(go.Scatter3d(
        x=H1_int, 
        y=H2_int, 
        z=H3_int,
        mode='markers',
        marker=dict(color=INTERSECTION_COLOR, size=INTERSECTION_SIZE),
        name='Intersection'
    ))

# 4E. (Optional) Axis lines for reference
max_extent = max(H_mag, A1, A2, A3) * 1.2
fig.add_trace(go.Scatter3d(
    x=[0, max_extent], y=[0, 0], z=[0, 0],
    mode='lines',
    line=dict(color='gray', width=3),
    name='H1-axis'
))
fig.add_trace(go.Scatter3d(
    x=[0, 0], y=[0, max_extent], z=[0, 0],
    mode='lines',
    line=dict(color='gray', width=3),
    name='H2-axis'
))
fig.add_trace(go.Scatter3d(
    x=[0, 0], y=[0, 0], z=[0, max_extent],
    mode='lines',
    line=dict(color='gray', width=3),
    name='H3-axis'
))

# -----------------------------------------------------------------------------
# 5. Update layout for better visuals
# -----------------------------------------------------------------------------
fig.update_layout(
    title='Momentum Sphere & Energy Ellipsoid: Colored Surface + Wireframe',
    scene=dict(
        xaxis_title='H₁',
        yaxis_title='H₂',
        zaxis_title='H₃',
        aspectmode='data'
    )
)

fig.show()
```
