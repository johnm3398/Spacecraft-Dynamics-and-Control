```python
A = [[1, 2], 
     [9, 10]]

part2 = " ".join(str(A[i][j]) for i in range(len(A)) for j in range(len(A[0])))
print(part2)

with open("module1_part2.txt", "w", encoding="utf-8") as f:
    f.write(part2)
```
