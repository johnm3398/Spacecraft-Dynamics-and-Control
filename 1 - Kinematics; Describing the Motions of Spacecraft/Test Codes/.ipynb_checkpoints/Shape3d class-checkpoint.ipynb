```python
import numpy as np
import plotly.graph_objects as go

class Shape3D:
    def __init__(self, shape_type, **params):
        self.shape_type = shape_type
        self.params = params
        self.vertices = None
        self.faces = None
        self.generate_shape()

    def generate_shape(self):
        if self.shape_type == 'cube':
            self.vertices, self.faces = self.generate_cube()
        elif self.shape_type == 'sphere':
            self.vertices, self.faces = self.generate_sphere()
        elif self.shape_type == 'pyramid':
            self.vertices, self.faces = self.generate_pyramid()
        elif self.shape_type == 'torus_with_arrow':
            self.vertices, self.faces = self.generate_torus_with_arrow()
        elif self.shape_type == 'gimbals':
            self.vertices = self.generate_gimbals()
            self.faces = None  # Faces are not used for gimbals visualization
        else:
            raise ValueError("Unsupported shape type")

    def generate_cube(self):
        # Define vertices and faces for a cube
        pass

    def generate_sphere(self):
        # Define vertices and faces for a sphere
        pass

    def generate_pyramid(self):
        # Define vertices and faces for a pyramid
        pass

    def generate_torus_with_arrow(self):
        # Define vertices and faces for a torus with an arrow
        pass

    def generate_gimbals(self):
        # Generate vertices for gimbals
        radius = self.params.get('radius', 1)
        thickness = self.params.get('thickness', 0.05)
        ring_points = 100

        # Generate a single ring
        theta = np.linspace(0, 2 * np.pi, ring_points)
        x_ring = radius * np.cos(theta)
        y_ring = radius * np.sin(theta)
        z_ring = np.zeros_like(theta)
        
        ring1 = np.vstack((x_ring, y_ring, z_ring)).T
        ring2 = np.vstack((x_ring, z_ring, y_ring)).T
        ring3 = np.vstack((z_ring, x_ring, y_ring)).T

        return [ring1, ring2, ring3]

    def plot(self):
        fig = go.Figure()

        if self.shape_type == 'gimbals':
            for ring in self.vertices:
                fig.add_trace(go.Scatter3d(x=ring[:, 0], y=ring[:, 1], z=ring[:, 2], mode='lines'))

        fig.update_layout(scene=dict(aspectmode='data'))
        fig.show()

# Example usage:
gimbals = Shape3D(shape_type='gimbals', radius=1, thickness=0.05)
gimbals.plot()


```
